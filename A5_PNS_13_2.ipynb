1ST Qustion

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N =69
calls = np.random.randint(0, 8, N)

unique, counts = np.unique(calls, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x=unique, y=pmf)
plt.title("PMF of Daily Calls")
plt.xlabel("Number of Calls")
plt.ylabel("Probability")

plt.subplot(1, 2, 2)
plt.plot(unique, cdf, marker='o', linestyle='-', color='b')
plt.title("CDF of Daily Calls")
plt.xlabel("Number of Calls")
plt.ylabel("Cumulative Probability")
plt.grid(True)

plt.tight_layout()
plt.show()
```

2ND Qestion

```python
from scipy.stats import binom

N = 100
p = 0.25
prob_20 = binom.pmf(20, N, p)
prob_at_least_30 = 1 - binom.cdf(29, N, p)
expected_patients = N * p

print(f"Probability of exactly 20 patients needing surgery: {prob_20:.4f}")
print(f"Probability of at least 30 patients needing surgery: {prob_at_least_30:.4f}")
print(f"Expected number of patients needing surgery: {expected_patients}")
```
